```python
import matplotlib.pyplot as plt
import numpy as np
filename='fig_'

import os
image_path = os.path.abspath(filename) # Figures out the absolute path for you in case your working directory moves around.
type(image_path)

data=np.loadtxt(fname='uxD2Q17couette.csv',delimiter = ',')

# data=np.loadtxt(fname='uxD2Q17couette.txt',delimiter = ',')
# data=np.loadtxt(fname='uxD2Q17couette.txt')
x=[]
y=[]
# z=[]

for i in range(len(data)):
		for j in range(4):
	 		if j==0:
	 			y.append(data[i][j])
			if j==2:
	 			x.append(data[i][j])
	 		# if j==2:
	 			# z.append(data[i][j])

for i in range(len(data)):
	x[i] = x[i] + 0.075

x1 = np.linspace(0,2.5,11)
x2 = np.linspace(0,0.0215131,11)

fig = plt.figure(figsize=(6,4))
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y,label='D2Q17')
ax1.plot(x1,x2,'bo',label = 'analytical')




ax1.set_xlabel('y' )
ax1.set_ylabel( 'ux-Magnitude')
plt.title('D2Q17 Ux vs y couette')

# # ax1.scatter(x,y,marker='x',s=14)
# # ax1.scatter(x,z,marker='*',s=14)


# # ax1.scatter(x,y,marker='x',s=0.2,label='Forward')
# # ax1.scatter(x1,y1,s=0.4,marker='*',label='Kevrekidis Projective')


ax1.grid(alpha=0.5,lw=0.5,ls='--',color='tab:gray')
ax1.legend()
plt.savefig(image_path+' D2Q17 Ux vs y couette.pdf', format='pdf', bbox_inches='tight')
```
