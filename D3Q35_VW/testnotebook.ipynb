```python
import matplotlib.pyplot as plt
import numpy as np
filename='fig_'

import os
image_path = os.path.abspath(filename) # Figures out the absolute path for you in case your working directory moves around.
type(image_path)

# data=np.loadtxt(fname='Q21_DRTT.txt',delimiter = ',',skiprows=1)





d1=np.loadtxt(fname='./datas/spurious_cur_w_o_correction.txt',delimiter =',',skiprows=1)
d2=np.loadtxt(fname='./datas/spurious_cur_w_correction.txt',delimiter =',',skiprows=1)





plt.figure()
plt.semilogy(d1[:,0],d1[:,1],'>',label = r'$\kappa_1 = \kappa $')
plt.semilogy(d2[:,0],d2[:,1],'<',label = r'$\kappa_1 = \kappa - correction $')
plt.grid(True,which='both',linewidth = 0.5,ls = '--')



plt.title("Spurious Currents in Bubble_VW")


plt.xlabel(r'$\theta/ \theta_c$')
plt.ylabel(r'$|u|^max$',)


# 

plt.legend()
plt.show()
```
